### Going further in the selection: isin()
The function **isin()** allows the selection of specific rows.

```python
energy_dec = energy[energy["Date"].isin(["2020-12-01", "2020-12-25"])]
```

### Another way to do the same thing without isin():

```python
energy_dec_bis = energy[(energy["Date"] == "2020-12-01") | (energy["Date"] == "2020-12-25")]
```

### Adding a column, creating a **datetime object** & deleting columns
Datetime is an object in python

0. (Optional) to_datetime('today')
1. Create a new column which combines the data in the columns "Date" and "Time" --> energy["DateTime"]
2. Convert this new column into a **datetime object** --> pd.to_datetime(energy["DateTime"])
3. (Optional) Remove the columns "Date" and "Time" from the dataframe --> energy.drop(["Date","Time"], axis = 1)

```python
pd.to_datetime("today")
```

```python
energy["DateTime"] = energy["Date"] + " " + energy["Time"]
```

```python
energy.head()
```

```python
energy["DateTime"] = pd.to_datetime(energy["DateTime"])
```

```python
energy.head()
```

```python
energy = energy.drop(["Date", "Time"], axis =1)
#axis = 0 --> index
#axis = 1 --> column
#axis is 0 by default.
```

```python
energy.head()
```

### Sort values in aquitaine
**sort_values vs. sort_index**

To use sort_values, we need to add a column name. To use sort_index, we just look at the column used as an index. Here our index column is DateTime. We specified this when using pd.read_csv("chart.csv", index_col = "DateTime"). When the index parameter is not defined, the existing index (row labels) is used. By default, the ascending is True for both cases


```python
aquitaine.sort_values(by = 'Nuclear')
```

### Set the column "Year" as index: set_index()

```python
energy_year = energy_year.set_index("Year")
```

### Select two columns to look at: pivot()
In the absence of an index column, then each row has 1 value and the rest are NaN. Each row is labeled 1, 2, 3, etc. With the year as index, each row corresponds to a year.
In the case of pivot(), the data is only rearranged. When multiple values need to be aggregated pivot_table() can be used, providing an aggregation function (e.g. mean) on how to combine these values.

```python
territory = energy_year.pivot(columns = "Territory", values = "Nuclear")
```

```python
territory.head(10)
```
